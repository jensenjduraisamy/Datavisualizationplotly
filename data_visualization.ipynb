```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'african_crises.csv'
data = pd.read_csv(file_path)

# Filter columns
kolumner_vi_behÃ¥ller = ["country", "year", "inflation_annual_cpi", "systemic_crisis", "exch_usd"]
Filtrerad_Grunddata = data[kolumner_vi_behÃ¥ller]

# Specify Zimbabwe
Namn = "Zimbabwe"
Refined_data = Filtrerad_Grunddata[Filtrerad_Grunddata["country"] == Namn]

# Specify years
star_year = 1994
end_year = 2008

# Filter by years
Refined_data = Refined_data[(Refined_data["year"] >= star_year) & (Refined_data["year"] <= end_year)]

# Create figure and axis
plt.figure(figsize=(10, 6))

# Define bar width
bar_width = 0.4

# Create bars for inflation
plt.bar(
    Refined_data["year"] - bar_width/2,  # Shift the bars to the left
    Refined_data["inflation_annual_cpi"],
    width=bar_width,
    color='skyblue',
    edgecolor='black',
    label='Inflation'
)

# Create bars for exchanged USD
plt.bar(
    Refined_data["year"] + bar_width/2,  # Shift the bars to the right
    Refined_data["exch_usd"] * 1000, 
    width=bar_width,
    color='orange',
    edgecolor='black',
    label='Exchanged USD (X 1000)'
)

# Use symlog scale for the y-axis with a threshold
threshold = 1
plt.yscale('symlog', linthresh=threshold)

# Labels and title
plt.xlabel('Ãr')
plt.ylabel('Value (Symlog scale)')
plt.title(f'Ãrlig inflation och vÃ¤xlade USD fÃ¶r {Namn} ({star_year}-{end_year})')
plt.xticks(rotation=45)

# Add annotation to indicate where the scale changes
plt.axhline(y=threshold, color='red', linestyle='--', label=f'Switch to Logarithmic Scale at {threshold}')

# Grid and layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Add legend
plt.legend()

# Show the plot
plt.show()
```
